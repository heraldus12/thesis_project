# RMSE Selection

```python
df = pd.read_excel('RMSE_F_selection.xlsx', index_col=0)

plt.figure(figsize=(10, 5))  # Set the figure size

# Plot each row in the DataFrame
for method in df.index:
    plt.plot(df.columns, df.loc[method], label=method)

# Customize the plot
plt.title('Feature selection Methods')
plt.xlabel('Number of Bandwidths')
plt.ylabel('RMSE')
plt.xticks(df.columns)  # Ensure all x-axis labels are shown
plt.legend()  # Show the legend
plt.grid()
plt.savefig('RMSE_F_selection.png', format='png', dpi=300, bbox_inches='tight')
plt.show()  # Display the plot
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the DataFrame
df = pd.read_csv("soilmoisture_dataset.csv", index_col=0)

# Remove empty rows
df = df.dropna()

# Reset the index without adding the old index as a column
df.reset_index(drop=True, inplace=True)

# Now proceed with feature selection
X = df.drop(['soil_temperature', 'datetime'], axis=1)
y = df['soil_moisture']

print(y.argmin())
print(y.argmax())


new_df = X.iloc[[y.argmin(), y.argmax()]].copy()
new_df.set_index('soil_moisture', inplace=True)
```

```python
import matplotlib.pyplot as plt

# Extract the two rows with minimum and maximum soil moisture
min_row = new_df.iloc[0]
max_row = new_df.iloc[1]

# Plotting the two rows
plt.figure(figsize=(9, 4))
plt.plot(min_row.index, min_row.values, label=f'Min SMC in % ({min_row.name})')
plt.plot(max_row.index, max_row.values, label=f'Max SMC in % ({max_row.name})')

# Customize the ticks: Select a subset of features to display
ticks_to_show = min_row.index[::4]  # Choose every 2nd feature (adjust as needed)
plt.xticks(ticks_to_show, rotation=45)  # Apply the selected ticks to the x-axis

# Adding labels and title
plt.xlabel('Bandwidths (nm)')
plt.ylabel('Hyperspectral reflectance')
plt.title('Comparison of bandwidth reflectancess at Min and Max Soil Moisture Content')
plt.legend()

# Show the plot
plt.tight_layout()
plt.savefig("EDA2.png")
plt.show()



```
